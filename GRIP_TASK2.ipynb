# Applying KMeans 

**ELBOW METHOD** for optimal value of K 

```python

SSE=[]
for i in range(1,10):
    kmeans=KMeans(n_jobs=-1,n_clusters=i,init='k-means++')
    kmeans.fit(df.iloc[:,[0,1,2,3]])
    SSE.append(kmeans.inertia_)
```

```python

df_1=pd.DataFrame({'Cluster':range(1,10),'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(df_1['Cluster'],df_1['SSE'],marker='*')
plt.xlabel("No. of Clusters")
plt.ylabel("Inertia")
plt.title("optimal value of k")
```

```python
#initalize k mean with 3 cluster
kmeans=KMeans(n_jobs=-1,n_clusters=3,init="k-means++")
kmeans.fit(df.iloc[:,[0,1,2,3]])
kmeans.cluster_centers_ 
```

```python
kmeans.labels_  #labels assign to each of record
```

```python
#putting this label in dataframe under cluster column to see that iris setosa given to us and 0 is predicted by model
df['cluster']=kmeans.labels_
df
```

```python
df['cluster'].unique()
```

```python
count_cluster=df.cluster.value_counts()
count_cluster

# Not 100%accurate
```

# VISUALIZE CLUSTER

```python
#3 cluster and black dot is centroid

plt.figure(figsize=(10,5))
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], c=df.cluster)
plt.title('Predicted Clusters\n')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1] ,  s=250, c='black', label = 'Centroids')
plt.show()
```
